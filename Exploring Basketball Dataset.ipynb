# Creating our own Function to be used for Plotting

```python
def myplot(data, player=Players):
    for name in player:
        plt.plot(data[Pdict[name]], ls='--',marker='o', ms=6, label=name)
    plt.xticks(list(range(0,10)), Seasons, rotation=45)
    plt.legend(Players, loc= 'upper left', bbox_to_anchor=(1,1))
    plt.rcParams['figure.figsize']= 12,6
    plt.title('Comparison')
    plt.show()

# The inputs taken by this function are the matrix (data) we want to visualize and players list by default. The working is simple, for
# every player in the Player list, the function will plot a point by looking up the value assigned to the key (player name)
# in the dictionary and the same value will correspond to that in the matrix we want to visualize. Other properties like ls,
# marker, ms, label, etc. are styling options for our plot
```

Now you must be thinking why there's a huge difference in the stats here. Don't Judge a Book by its Cover they say. So that's what we are going to drill down into next. The Player Style of each individual will help us get some valuable insight.

```python
myplot((Points-FreeThrows) / FieldGoals)
```

Okay. So what does this all mean? Let's have a little backgroung info on points system in BasketBall. I'm not going to keep it simple and short, the greater the points, the further the distance from where the ball has been shot from. That sounds simple, right? So, a 2-Pointer represents a distance closer to the hoop and a 3-Pointer represents a distance upto the half court.

Now that we're clear on this, let's study the above viz and try to get some meaning out of it. Dwigh Howard 2-points consistency shows that he isn't a far-shooter type. However, some players like Joe Johnson and Chris Bosh have really switched up their game and gone beyond 2.2 points on average, indicating that they're more than capable of holding their own from the middle of the court!
